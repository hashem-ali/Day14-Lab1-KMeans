# EDA

## Create a scatterplot of Grad.Rate versus Room.Board where the points are colored by the Private column. 

```python
plt.figure(figsize=(14,9))
sns.scatterplot(data= df, x = df['Room.Board'], y = df['Grad.Rate'], hue=df['Private']);
plt.legend()
plt.grid()
plt.tick_params(top=True, bottom=True, left=False, right='off', labelleft='off', labelbottom='on')
#plt.box(False)
```

## Create a scatterplot of F.Undergrad versus Outstate where the points are colored by the Private column.

```python
plt.figure(figsize=(14,9))
sns.scatterplot(data=df, x=df['Outstate'], y=df['F.Undergrad'], hue=df['Private'], alpha=0.3);
plt.grid()
# plt.tick_params(top='off', bottom='off', left='off', right='off', labelleft='off', labelbottom='on')
plt.box()
```

## Use sns.FacetGrid to create a stacked histogram showing Out of State Tuition based on the Private column. 

```python
sns.FacetGrid(df, size=10).map_dataframe(sns.histplot, x="Outstate", hue="Private");
plt.grid()
plt.box()
```

## Use sns.FacetGrid to create a stacked histogram showing Grad.Rate based on the Private column. 

```python
sns.FacetGrid(df, size=10).map_dataframe(sns.histplot, x="Grad.Rate", hue="Private");
plt.grid()
plt.box()
```

## Notice how there seems to be a private school with a graduation rate of higher than 100%.What is the name of that school?

```python
filt = (df['Private'] == 'Yes') & (df['Grad.Rate'] > 100)
df[filt]
```

## Set that school's graduation rate to 100 then re-do the histogram visualization.

```python
df.loc[filt, 'Grad.Rate'] = 100
```

```python
df[filt]
```

```python
sns.FacetGrid(df, size=10).map_dataframe(sns.histplot, x="Grad.Rate", hue="Private");
plt.grid()
plt.box()
```

## K Means Model

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=2)
```

## Fit the model to all the data except for the Private label.

```python
x = df.drop(['Private'], axis=1)
x
```

```python
kmeans.fit(x)
```

## What are the cluster center vectors?

```python
kmeans.cluster_centers_
```

## Create a new column for df called 'Cluster', which is a 1 for a Private school, and a 0 for a public school.

```python
def df_converter(cluster):
    if cluster == 'Yes':
        return 1
    else:
        return 0
df['Cluster'] = df['Private'].apply(df_converter)
df.head()
```

## Create a confusion matrix and classification report to see how well the Kmeans clustering worked without being given any labels.

```python
from sklearn.metrics import confusion_matrix, classification_report
print(confusion_matrix(df["Cluster"], kmeans.labels_))
```

```python
print(classification_report(df["Cluster"], kmeans.labels_))
```
